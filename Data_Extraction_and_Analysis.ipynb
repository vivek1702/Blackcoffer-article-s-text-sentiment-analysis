Extract the data from file using beautiful soap and stored in root dir

```python
from nltk.corpus.reader import titles

for index, row in df.iterrows():
  url_id = row['URL_ID']
  url = row['URL']

  header={"User-Agent": "Mozilla/5.0 (Linux; Android 11; SAMSUNG SM-G973U) AppleWebKit/537.36 (KHTML, like Gecko) SamsungBrowser/14.2 Chrome/87.0.4280.141 Mobile Safari/537.36"}

  try:
    response = requests.get(url, headers=header)
  except:
    print("Cannot get response from {}".format(url_id))

  try:
    soup = BeautifulSoup(response.content, 'html.parser')
  except:
    print("Can not get page of {} ".format(url_id))

  try:
    title = soup.find('h1').get_text()
  except:
    print("cannot find the title of article {}".format(url_id))
    continue

  Article = ""
  try:
    for p in soup.find_all('p'):
      Article += p.get_text()
  except:
    print("can't find text for Article {}".format(url_id))

  file_name = root_dir+f"{url_id}.txt"
  with open(file_name, 'w') as file:
    file.write(title + '\n' + Article)

```

```python
#Directory of all the files
text_dir = "/content/gdrive/My Drive/20211030 Test Assignment/Extracted_Data/"
stopwords_dir = "/content/gdrive/My Drive/20211030 Test Assignment/StopWords/"
sentiment_dir = "/content/gdrive/My Drive/20211030 Test Assignment/MasterDictionary/"
```

Load all the files from stopwords and stroe then in set variables

```python
stop_words = set()

for files in os.listdir(stopwords_dir):
    with open(os.path.join(stopwords_dir, files),'r',encoding='ISO-8859-1') as f:
        stop_words.update(set(f.read().splitlines()))
```

```python
stop_words
```

load all the extracted article data and stored them into variable

```python
docs = []
for files in os.listdir(text_dir):
    with open(os.path.join(text_dir, files), 'r', encoding='ISO-8859-1') as f:
        text = f.read()

        #will tokenize the word and check for word which is not in stop word
        words = word_tokenize(text)
        filter_words = []
        for word in words:
            if word.lower() not in stop_words:
                filter_words.append(word)

        docs.append(filter_words)
```

```python
print(docs)
```

now will read the postive and negative text file and store them into set variable

```python
pos = set()
neg = set()

for file in os.listdir(sentiment_dir):
    if file == 'positive-words.txt':
        with open(os.path.join(sentiment_dir,file), 'r', encoding='ISO-8859-1') as f:
            pos.update(set(f.read().splitlines()))
    elif file == 'negative-words.txt':
        with open(os.path.join(sentiment_dir,file), 'r', encoding='ISO-8859-1') as f:
            neg.update(set(f.read().splitlines()))
```

```python
print(pos)
```

Now will check postive score and negative score plus will calculate polarity score, subjectve score

```python
positive_words = []
negative_words = []

for i in range(len(docs)):
    for word in docs[i]:
        if word.lower() in pos:
            positive_words.append(word)

    for word in docs[i]:
        if word.lower() in neg:
            negative_words.append(word)
```

```python
positive_score = []
negative_score = []
polarity_score = []
subjective_score = []

for i in range(len(docs)):
    positive_score.append(len(positive_words[i]))
    negative_score.append(len(negative_words[i]))
    polarity_score.append((positive_score[i]- negative_score[i])/(positive_score[i] + negative_score[i]) + (0.000001))
    subjective_score.append((positive_score[i]- negative_score[i])/(len(docs[i])) + (0.000001))
```

Analysis of readability

*   Average Sentence Length = the number of words / the number of sentences
*   Percentage of Complex words = the number of complex words / the number of words

*  Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)







```python
avg_sentence_length = []
Percentage_of_Complex_words  =  []
Fog_Index = []
complex_word_count =  []
avg_syllable_word_count = []

stop_words = set(stopwords.words('english'))

def measures(file):
    with open(os.path.join(text_dir, file), 'r') as f:
        text = f.read()

    #remove puncutations
    text = re.sub(r'[^\w\s.]', '', text)

    #split the given file into sentences
    sentences = text.split('.')
    num_sentences = len(sentences)

    #total words in file
    words = []
    for word in text.split():
        if word.lower() not in stop_words:
            words.append(word)

    #number of words
    num_words = len(words)

    #complex words are the words contains more than two syllables
    complex_words = []
    for word in words:
        vowels = 'aeiou'
        Syllables_word_count = sum(1 for letter in word if letter.lower() in vowels)
        if Syllables_word_count > 2:
            complex_words.append(word)

    #syllables count each word
    #we count number of syllabes in each word of the text by counting the number of vowels

    syllables_count = 0
    syllables_word = []
    for word in words:
        if word.endswith('es') or word.endswith('ed'):
            word[:-2]

        vowels = 'aeiou'
        syllable_count_word = sum( 1 for letter in word if letter.lower() in vowels)
        if syllable_count_word >= 1:
            syllables_word.append(word)
            syllables_count += syllable_count_word

    avg_sentence_len = num_words / num_sentences
    avg_syllable_word_count = syllables_count / len(syllables_word)
    Percent_Complex_words  =  len(complex_words) / num_words
    Fog_Index = 0.4 * (avg_sentence_len + Percent_Complex_words)

    return avg_sentence_len, Percent_Complex_words, Fog_Index, len(complex_words),avg_syllable_word_count
```

```python
# iterate through each file or doc
for file in os.listdir(text_dir):
  x,y,z,a,b = measures(file)
  avg_sentence_length.append(x)
  Percentage_of_Complex_words.append(y)
  Fog_Index.append(z)
  complex_word_count.append(a)
  avg_syllable_word_count.append(b)
```

```python
def cleaned_words(file):
  with open(os.path.join(text_dir,file), 'r') as f:
    text = f.read()
    text = re.sub(r'[^\w\s]', '' , text)
    words = [word  for word in text.split() if word.lower() not in stop_words]
    length = sum(len(word) for word in words)
    average_word_length = length / len(words)
  return len(words),average_word_length

word_count = []
average_word_length = []
for file in os.listdir(text_dir):
  x, y = cleaned_words(file)
  word_count.append(x)
  average_word_length.append(y)
```

```python
# To calculate Personal Pronouns mentioned in the text, we use regex to find
# the counts of the words - “I,” “we,” “my,” “ours,” and “us”. Special care is taken
#  so that the country name US is not included in the list.
def count_personal_pronouns(file):
  with open(os.path.join(text_dir,file), 'r') as f:
    text = f.read()
    personal_pronouns = ["I", "we", "my", "ours", "us"]
    count = 0
    for pronoun in personal_pronouns:
      count += len(re.findall(r"\b" + pronoun + r"\b", text)) # \b is used to match word boundaries
  return count

pp_count = []
for file in os.listdir(text_dir):
  x = count_personal_pronouns(file)
  pp_count.append(x)
```
